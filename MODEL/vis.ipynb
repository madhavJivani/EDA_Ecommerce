```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['discounted_mrp'])
plt.title('Box plot of Discounted MRP')
plt.xlabel('Discounted MRP')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['color'], kde=False)
plt.title('Histogram of Colors')
plt.xlabel('Color')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['discounted_mrp'].quantile(0.25)
Q3 = df['discounted_mrp'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Print the results
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# # Plot the boxplot with IQR
# plt.figure(figsize=(12, 6))
# sns.boxplot(x=df['discounted_mrp'])
# plt.axvline(lower_bound, color='r', linestyle='--', label='Lower Bound')
# plt.axvline(upper_bound, color='g', linestyle='--', label='Upper Bound')
# plt.title('Box plot of Discounted MRP with IQR')
# plt.xlabel('Discounted MRP')
# plt.legend()
# plt.show()
# Plot the histogram with KDE and outliers band
plt.figure(figsize=(12, 6))
sns.histplot(df['discounted_mrp'], kde=True)
plt.axvline(lower_bound, color='r', linestyle='--', label='Lower Bound')
plt.axvline(upper_bound, color='g', linestyle='--', label='Upper Bound')
plt.title('Histogram of Discounted MRP with KDE and Outliers Band')
plt.xlabel('Discounted MRP')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
